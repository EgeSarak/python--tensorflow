```python
import pandas as pd
```

```python
df=pd.read_excel("bisiklet_fiyatlari.xlsx")
```

```python
df.head()
```

```python
sbn.pairplot(df);
```

```python
from sklearn.model_selection import train_test_split
```

```python
df
```

```python
y=df["Fiyat"].values
X=df[["BisikletOzellik1","BisikletOzellik2"]].values
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=15)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
 #scaling
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential #modelimizi bu sınıftan olusutruyoruz,ondan sonra içersinde hangi katmanlarla çalısıcaz onu belirliyoruz
from tensorflow.keras.layers import Dense 
```

```python
model=Sequential()

model.add(Dense(4,activation="relu")) #4 tane nöron olacak,aktivasyonda relu olacak
model.add(Dense(4,activation="relu"))
model.add(Dense(4,activation="relu"))

model.add(Dense(1)) #1 tane nöron verdik,çıktı katmanı

model.compile(optimizer="rmsprop",loss="mse") #derleme işlemi

#model hazırlandı
```

```python
model.fit(X_train_scaled,y_train,epochs=250)
```

```python
#loss un degerleri sözlük içinde veriyor
loss=model.history.history["loss"]  #dizi haline getirdik
```

```python
sbn.lineplot(x=range(len(loss)),y=loss);
#çizgi grafiği olustruduk,x ekseninde loss içersinde kaç tane veri varsa onu göstersin,y ekseninde de loss un kendisi olsun
```

```python
#modelin kendi değerlendirmesi için evaluate
trainLoss=model.evaluate(X_train_scaled,y_train,verbose=0) #verbose 0 ile deirekt sonucu gördük detayları istemedik.
```

```python
testloss=model.evaluate(X_test_scaled,y_test,verbose=0)
```

```python
y=df["Type"].values
X=df.drop("Type",axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=15)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation,Dropout
from tensorflow.keras.callbacks import EarlyStopping #overfittingi çözmek için kullanıcaz.droputu da aynı amaçla.
```

```python
X_train_scaled.shape
```

```python
model=Sequential()

model.add(Dense(units=30,activation="relu"))#giriş katmanı #units-->30 tane içeri nöron koy demek
#30 olmasının nedeni;kaç kolon varsa giriş değerimiziona eşit yapmamız önerilir.

model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=1,activation="sigmoid")) #cıkıs katmanı

model.compile(loss="binary_crossentropy",optimizer="adam")   #derleme
```

```python
model.fit(X_train_scaled,y_train,epochs=700,validation_data=(X_test_scaled,y_test),verbose=1)
```

```python
model=Sequential()

model.add(Dense(units=30,activation="relu"))#giriş katmanı #units-->30 tane içeri nöron koy demek
#30 olmasının nedeni;kaç kolon varsa giriş değerimiziona eşit yapmamız önerilir.

model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=1,activation="sigmoid")) #cıkıs katmanı

model.compile(loss="binary_crossentropy",optimizer="adam")   #derleme
```

```python
earlystopping=EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25)
```

```python
model.fit(X_train_scaled,y_train,epochs=700,validation_data=(X_test_scaled,y_test),verbose=1,callbacks=[earlystopping])
```

```python
modelKaybi=pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot()
```

```python
model=Sequential()

model.add(Dense(units=30,activation="relu"))#giriş katmanı #units-->30 tane içeri nöron koy demek
#30 olmasının nedeni;kaç kolon varsa giriş değerimiziona eşit yapmamız önerilir.
model.add(Dropout(0.6)) #genelde 0.50

model.add(Dense(units=15,activation="relu"))
model.add(Dropout(0.6)) #genelde 0.50

model.add(Dense(units=15,activation="relu"))
model.add(Dropout(0.6)) #genelde 0.50

model.add(Dense(units=1,activation="sigmoid")) #cıkıs katmanı

model.compile(loss="binary_crossentropy",optimizer="adam")   #derleme
```

```python
model.fit(X_train_scaled,y_train,epochs=700,validation_data=(X_test_scaled,y_test),verbose=1,callbacks=[earlystopping])
```

```python
kayipDf=pd.DataFrame(model.history.history)
```

```python
kayipDf.plot()
```

```python
tahminlerimiz= np.argmax(model.predict(X_test_scaled),axis=1)
```

```python
tahminlerimiz
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,tahminlerimiz))
```

```python
print(confusion_matrix(y_test,tahminlerimiz))
```
